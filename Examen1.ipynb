## 1.- Considere la función f(x) = $e^{-x}- ln(x)$, a partir de esta analice la ecuación $f(x)=0$ y determine una raíz de la misma con al menos una precisión de $10^{-10}$ empleando el método de bisección.

### Definición del intervalo de busqueda:
<img src=" ">

#### Desde la graficación de la función, podemos observar que recorre desde un punto inicial 0 hasta aproximadamente un punto 1.3. Entonces definimos sus extremos en [0,1.3]

```python
# Importaremos las funciones a emplear necesarias para definir la ecuación algebraíca
import numpy as np
import math as math

# Condiciones de conlusion del método
tol = 1.*(10**(-10)) # Obtener un error relativo de 10^(-10)
maxItera = 100 # Realizar máximo 100 iteraciones

# Definimos la funcion objetivo
def fx(x):
    fx = (np.exp(-x))-np.log(x)
    return fx

# Definimos el intervalo de busqueda inicial y validamos que funcione para el método
a = 0 # Extremo izquierdo
b = 1.31 # Extremo derecho
k = 0 # Inicilizamos las iteraciones
error = 10000 # Inicializamos el valor del error en un número muy grande

# Definimos el método iterativo
def itebiseccion(a,b,fx):
    
    # Calculamos el punto medio
    pmedio = a + 0.5*(b-a)
    
    # Validamos que intervalo podemos emplear
    if fx(a)*fx(pmedio)<0:
        # Mandamos una bandera que aún no se ha encontrada la raíz y redefinimos b = punto medio
        return 1, a, pmedio, pmedio
    
    elif fx(pmedio)*fx(b)<0:
        # Mandamos una bandera que aún no se ha encontrada la raíz y redefinimos a = punto medio
        return 1, pmedio, b, pmedio
    
    else:
        # Mandamos una bandera que se ha encontrada la raíz y regresamos la raíz
        return 0, pmedio
```

```python
# Empezamos el método de bisección verificando las diferentes condiciones sobre el intervalo para lo que se
# validan las posibilidades de la existencia de la raíz en el intervalo dado

# En este caso no existe una raíz en el intervalo
if fx(a)*fx(b)>0:
    # Se describe que no existe una raíz en el intervalo dado
    print('No existen raices de la funcion objetivo en el intervalo dado \n')

# En este caso se ha encontrado la raíz y esta puede estar en a o en b.
elif fx(a)*fx(b)==0:
    
    # La raíz se encuentra en el punto a
    if fx(a)==0:
        # Se describe que la raíz se encuentra en el punto a
        print('La raíz de la ecuación se encuentra en el punto: ', a)
        
    # Dado que la raíz no es a, entonces es b.
    else:
        # Se describe que la raíz se encuentra en el punto b
        print('La raíz de la ecuación se encuentra en el punto: ', b)

# Dado que existe una raíz en el intervalo comenzaremos el proceso iterativo
else:
    
    # Señalamos que iniciamos el proceso iterativo
    print('Iniciamos el proceso iterativo')
    
    # Iniciamos el proceso iterativo
    while (error > tol and k < maxItera):
        
        # Iteramos hasta satisfacer las condiciones o encontrar una raiz
        itera = itebiseccion(a,b,fx)
        
        # Validamos los casos obtenidos de la bandera, si esta es 0, se ha determinado la raíz
        if itera[0] == 0:
            # Describimos que hemos determinado la raíz y la imprimimos
            print('La raíz de la ecuación se encuentra en el punto: ', itera[1])
            
        # Validamos los casos obtenidos de la bandera, si esta es 1, no se ha determinado la raíz
        else:
            # Incrementamos en 1 las iteraciones realizadas
            k = k + 1
            
            # Imprimimos los resultados obtenidos de la iteracion
            print('Resultados de la iteracion: ', k)
            print('La aproximación de la raíz se encuentra dada por: ', itera[3])
            print('El valor de la función bajo la aproximación es: ', fx(itera[3]))
            print('El intervalor de búsqueda se ha reducido al [', itera[1], ' , ', itera[2], ' ]')
            print('El tamaño del intervalo donde se encuentra la raiz es ', itera[2]-itera[1], '\n')
            
            # Actualizamos las variables
            a = itera[1]
            b = itera[2]
            error = b - a
```

### Procedimiento

y lo debemos tranformar a un g(x). El cual será: $$g(x) = -e^{-x}+2 $$

Una vez encontrado sus iteraciones, debemos evaluar el resultado de la raiz en la f(x). Mientras más cerca se encuentre del valor 0.0, tendra un menor error

## 4.- Considere la función $f(x)=x^2-2cos(x)+1$, a partir de esta analice la ecuación $f(x)=0$ y determine la raíz por medio del método de newton, se deberá de especificar la elección del punto incial.

### Procedimiento 
Consideramos la función $f(x) = x^2-2cos(x)+1$ a partir de la ecuación $f(x)=0$

Se considera $f´(x)= 2x + 2sen(x)$  y $f´´(x)= 2cos(x)+1 $

Tomando a $p0 = 0.5$, pues cumple con $f´(p0)≠0$:

$f´(0.5)=1+2sen(1/2) = 1.017$

```python
from numpy import cos, sin, exp

tol = 1.*(10**(-10)) #error relativo
maxItera = 100 # 100 iteraciones

def fx(x):
    fx = x**2- 2*cos(x) + 1
    return fx

def fxp(x):
    fxp = 2*(sin(x)+x)
    return fxp

def fxpp(x):
    fxpp = 2*(cos(x)+1) 
    return fxpp

p0 = 0.5
k = 1 
error = 10000 
pAnterior = p0 

def itenewtonmultiple(a, fx, fxp, fxpp):
    evalua = fx(a)

    if evalua==0:
        return 0, a
    else:
        return 1, a -((fx(a)*fxp(a))/(fxp(a)**2-fx(a)*fxpp(a)))
if fx(p0) == 0:
    print('Se ha encontrado la raíz de la ecuación, la cual está dado por el punto {0}.'.format(p0))

else:
    print('Iniciamos el proceso iterativo')
    
    while (error > tol and k < maxItera):
        
        itera = itenewtonmultiple(p0, fx, fxp, fxpp)
        
        if itera[0] == 0:
            print('La raíz de la ecuación se encuentra en el punto: {0}.'.format(itera[1]))
            k = maxItera
            
        else:
            k = k + 1
            
            print('Resultados de la iteracion {0}.'.format(k))
            print('La aproximación de la raíz se encuentra dada por {:.8}.'.format(itera[1]))
            print('El valor de la función bajo la aproximación es {:.8}.'.format(fx(itera[1])))
            
            error = abs(p0 - itera[1]) + abs(fx(p0)) # Error de aproximaciones
            p0 = itera[1]
            
            print('El error relativo se encuentra dado por {:.8}.\n'.format(error))
```

## 5.- Aplique el método de la secante para el ejercicio anterior y compare los resultados así como el número de iteraciones realizadas.

### Procedimiento 
Consideramos la función $f(x) = x^2-2cos(x)+1$ a partir de la ecuación $f(x)=0$ y se definen ambos puntos (0.5 y 0.75) por aproximaciones según su grafica: 

<img src=" ">


```python
from numpy import cos

tol = 1.*(10**(-20))
maxItera = 100
def fx(x):
    fx = x**2- 2*cos(x) + 1
    return fx
p0 = 0.5
p1 = 0.75
k = 0
error = p1 - p0
def itesecante(a, b, fx):
    evalua = fx(b)
    if evalua == 0:
        return 0, b
    else:
        return 1, b - fx(b)*(b-a)/(fx(b)-fx(a))
if fx(p1) == 0:
    print('Se ha encontrado la raíz de la ecuación, la cual está dado por el punto {0}.'.format(p1))
else:
    print('Iniciamos el proceso iterativo')
    while (error > tol and k < maxItera):
        itera = itesecante(p0, p1, fx)
        if itera[0] == 0:
            print('La raíz de la ecuación se encuentra en el punto: {:.8}.'.format(itera[1]))
            k = maxItera
        else:
            k = k + 1
            error = abs(p1 - itera[1]) + abs(fx(itera[1])) # Error de aproximaciones
            p0 = p1
            p1 = itera[1]
            print('Resultados de la iteracion {0}.'.format(k))
            print('La aproximación de la raíz se encuentra dada por {:.8}.'.format(itera[1]))
            print('El valor de la función bajo la aproximación es {:.8}.'.format(fx(itera[1])))
            print('El error relativo se encuentra dado por {:.8}.\n'.format(error))
```
